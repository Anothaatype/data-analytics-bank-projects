# Data Cleaning

###Price Area Columns

Process of Cleaning :

*   Changing The Data Type of Discounted Price
*   Changing The Data Type of Actual Price
*   Changing The Data Type of Discounted Percentage
*   Changing The Values of Percentage from % into Decimal
*   Showing The Results for each Discounted Price, Actual Price, Discounted Price





```python
#Changing the data type of discounted price actual price
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')
```

```python
#Changing the data type of Actual Price
df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
#Changing Datatype in Discount Percentage
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

#Changing Values in Discount Percentage
df['discount_percentage'] = df['discount_percentage'] / 100
df['discount_percentage']
```

```python
#Showing the output that already Changed

df[['discounted_price', 'actual_price', 'discount_percentage']]
```

#### Creating Categories Ranking based on Rating Score

Process of Creating Ranking in Categories by Rating Score :

"this sections focuses are for differentiate the ranking by the rating scorem in this terms i create it into 5 rank calculation terms"

*  Creating Categories For Rankings
*  Creating a Rating Score Column
*  Creating Different Price Column (Actual Price - Discount Price)
*  Showing The Result of Changing



##### Categories Ranking

```python
#Creating Categories for Rankings

rating_score = []

for score in df1['rating']:
    if score < 2.0 : rating_score.append('Poor')
    elif score < 3.0 : rating_score.append('Below Average')
    elif score < 4.0 : rating_score.append('Average')
    elif score < 5.0 : rating_score.append('Above Average')
    elif score == 5.0 : rating_score.append('Excellent')
```

##### Rating Score

```python
#Creating A new Column and Changing the Data Type

df1['rating_score'] = rating_score
df1['rating_score'] = df1['rating_score'].astype('category')
```

```python
#Reordered Categories

df1['rating_score'] = df1['rating_score'].cat.reorder_categories(['Below Average', 'Average', 'Above Average', 'Excellent'], ordered=True)
```

##### Difference Price (Actual Price-Discounted Price)

```python
#Creating Difference of Price Column between Actual Price and Discounted Price

df1['difference_price'] = df1['actual_price'] - df1['discounted_price']
```

### #**2** Creating New DataFrame With Selected Column - Reviewers and Users Section

Process of Creating Reviewers :

*   Identify which column belong to reviewers section (user_id and user_name)
*   Fixing the naming, data types, and values in each user_id and user_name
*   Creating User ID displaying only 1 ID per row
*   Merging the User_ID that already done in process 2 and User_Name that already done in process 2 into one DataFrame



```python
#Subsetting Reviewers Identifications

reviewers = df[['user_id','user_name']]
reviewers
```

##### Reviewers - User ID

```python
#Splitting the strings in user_id column

reviewer_id_split = reviewers['user_id'].str.split(',', expand=False)

reviewer_id_split
```

```python
#Making user id display 1 id per row

reviewer_id_exp = reviewer_id_split.explode()

reviewer_id_clean = reviewer_id_exp.reset_index(drop=True)

reviewer_id_clean
```

##### Reviewers - User Name

```python
#Splitting the strings in user_name column

reviewer_name_split = reviewers['user_name'].str.split(',', expand=False)

reviewer_name_split
```

```python
#Making user name display 1 id per row

review_name_exp = reviewer_name_split.explode()

reviewer_name_clean = review_name_exp.reset_index(drop=True)

reviewer_name_clean
```

#### Merging User ID and User Name

```python
#Creating 2 Data Frames to be merged

df21 = pd.DataFrame(data=reviewer_id_clean)
df22 = pd.DataFrame(data=reviewer_name_clean)
```

#### #2 Clean DataFrame Results Reviewers (User ID-User Name)

```python
#Merging the 2 dataframe containing user_id and user_name

df2 = pd.merge(df21, df22, left_index=True, right_index=True)
```

###New DataFrame Results | Copy

```python
#Creating Copy Clean DataSet
dfC = df1.copy() #Clean Dataset on Product Category
dfR = df2.copy() #Clean Dataset on Reviewer
```

```python
#Creating Main New Dataset
dfAll = pd.merge(dfC, dfR, left_index=True, right_index=True).copy()
```
